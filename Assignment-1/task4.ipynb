```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df_raw = pd.read_csv('ODI-2026.csv', sep=';', encoding='utf-8-sig')
```

Use the cleaned version from task 1b

```python
df_clean = df_raw.copy()

# Convert to numeric to detect outliers
df_clean['stress'] = pd.to_numeric(df_clean['What is your stress level (0-100)?'], errors='coerce')
df_clean['sports'] = pd.to_numeric(df_clean['How many hours per week do you do sports (in whole hours)? '], errors='coerce')

# Set invalid stress values to NaN
df_clean.loc[df_clean['stress'] > 100, 'stress'] = None

# Set invalid sports values to NaN
df_clean.loc[df_clean['sports'] < 0, 'sports'] = None

stress_nan_idx = df_clean[df_clean['stress'].isna()].index
sports_nan_idx = df_clean[df_clean['sports'].isna()].index

df_median = df_clean.copy()
df_median['stress'] = df_median['stress'].fillna(df_median['stress'].median())
df_median['sports'] = df_median['sports'].fillna(df_median['sports'].median())
```

```python

df = df_median.copy()

#print all columns
print(df.columns)

numerical_columns = ['How many students do you estimate there are in the room?',
                     'stress',
                     'sports']

categorical_columns = ['Have you taken a course on machine learning?',
        'Have you taken a course on information retrieval?',
        'Have you taken a course on statistics?',
        'Have you taken a course on databases?', 'What is your gender?',
        'I have used LLMs to help me with some of my study assignments ',]
```

```python
from sklearn.model_selection import train_test_split

feature_cols = numerical_columns

df['Give a random number'] = pd.to_numeric(
    df['Give a random number'].astype(str).str.replace(',', '.', regex=False),
    errors='coerce'
)

model_df = df[feature_cols + ['Give a random number']].copy()

for col in numerical_columns:
    model_df[col] = pd.to_numeric(model_df[col], errors='coerce')

model_df = model_df.dropna()

#see if there are any nans
print(model_df.isna().sum())

X_num = model_df[numerical_columns]
#X_cat = pd.get_dummies(model_df[categorical_columns].fillna('missing'), drop_first=True)
#X_model = pd.concat([X_num, X_cat], axis=1)
X_model = X_num
y = model_df['Give a random number']

X_train, X_test, y_train, y_test = train_test_split(X_model, y, test_size=0.25, random_state=42)

lr = LinearRegression()
lr.fit(X_train, y_train)

print('Train: R^2', lr.score(X_train, y_train))
print('Test R^2:', lr.score(X_test, y_test))
```
